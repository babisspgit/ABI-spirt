```python
import pandas as pd
import numpy as np
import datetime
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv("Trips_2018.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['starttime']=pd.to_datetime(df['starttime'], format="%Y-%m-%d %H:%M:%S.%f")
df['stoptime']=pd.to_datetime(df['stoptime'], format="%Y-%m-%d %H:%M:%S.%f")
```

```python
df.info()
```

```python
le = LabelEncoder()
```

```python
df['usertype'] = le.fit_transform(df['usertype'])
df = df.dropna(subset=['start_station_id', 'end_station_id'])
df['start_station_id'] = df['start_station_id'].astype(int)
df['end_station_id'] = df['end_station_id'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

# Step 1: Data Preparation
unique_stations = df.drop_duplicates(subset=['start_station_id', 'start_station_latitude', 'start_station_longitude'])
stations = unique_stations[['start_station_id', 'start_station_latitude', 'start_station_longitude']]

# Step 2: K-Means Clustering
num_clusters = 20
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(stations[['start_station_latitude', 'start_station_longitude']])

# Step 3: Assigning Cluster Labels
stations['cluster_id'] = kmeans.labels_

# Step 4: Visualization
labels = kmeans.labels_
centers = kmeans.cluster_centers_

plt.figure(figsize=(10, 10))

for i in range(num_clusters):
    cluster = stations[stations['cluster_id'] == i]
    plt.scatter(cluster['start_station_longitude'], cluster['start_station_latitude'], label=f'Cluster {i}')
    plt.scatter(centers[i, 1], centers[i, 0], marker='x', s=100, linewidths=3, color='k')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Station Clusters')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assuming df is your DataFrame
# Getting unique stations
unique_stations = df.drop_duplicates(subset=['start_station_id', 'start_station_latitude', 'start_station_longitude'])
stations = unique_stations[['start_station_id', 'start_station_latitude', 'start_station_longitude']]

# Number of clusters
num_clusters = 20

# Perform K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(stations[['start_station_latitude', 'start_station_longitude']])

# Assign cluster labels to each station
stations['cluster_id'] = kmeans.labels_

# Counting the number of stations in each cluster
cluster_counts = stations['cluster_id'].value_counts().reset_index()
cluster_counts.columns = ['Cluster', 'Number of Stations']

# Sorting clusters by cluster number
cluster_counts = cluster_counts.sort_values(by='Cluster').reset_index(drop=True)

# Creating a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(cluster_counts['Cluster'].astype(str), cluster_counts['Number of Stations'])

# Annotating each bar with the respective count
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')  # va: vertical alignment

plt.xlabel('Cluster')
plt.ylabel('Number of Stations')
plt.title('Number of Stations in Each Cluster')
plt.xticks(rotation=45)
plt.show()

```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Assuming df is your DataFrame
# Getting unique stations for start stations
unique_start_stations = df.drop_duplicates(subset=['start_station_id', 'start_station_latitude', 'start_station_longitude'])
start_stations = unique_start_stations[['start_station_id', 'start_station_latitude', 'start_station_longitude']]

# Getting unique stations for end stations
unique_end_stations = df.drop_duplicates(subset=['end_station_id', 'end_station_latitude', 'end_station_longitude'])
end_stations = unique_end_stations[['end_station_id', 'end_station_latitude', 'end_station_longitude']]

# Number of clusters
num_clusters = 20

# Perform K-Means clustering for start stations
kmeans_start = KMeans(n_clusters=num_clusters, random_state=0).fit(start_stations[['start_station_latitude', 'start_station_longitude']])
start_stations['start_cluster_id'] = kmeans_start.labels_

# Perform K-Means clustering for end stations
kmeans_end = KMeans(n_clusters=num_clusters, random_state=0).fit(end_stations[['end_station_latitude', 'end_station_longitude']])
end_stations['end_cluster_id'] = kmeans_end.labels_

# Creating mappings
start_station_cluster_mapping = start_stations.set_index('start_station_id')['start_cluster_id'].to_dict()
end_station_cluster_mapping = end_stations.set_index('end_station_id')['end_cluster_id'].to_dict()

# Creating df2
df2 = df.copy()
df2['start_cluster_id'] = df2['start_station_id'].map(start_station_cluster_mapping)
df2['end_cluster_id'] = df2['end_station_id'].map(end_station_cluster_mapping)
df2 = df2[['starttime', 'stoptime', 'start_station_id', 'start_cluster_id', 'end_station_id', 'end_cluster_id']]
```
